import pandas as pd import matplotlib.pyplot as plt import seaborn as
sns

df =
pd.read_csv(‘https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv’)
\# Descriptive Analysis print(df.head()) print(f”Number of rows:
{df.shape\[0\]}“) print(f”Number of columns: {df.shape\[1\]}“)
print(df.info()) print(df.describe()) \# Data Exploration print(f”Number
of unique Campaign_ID values: {df\[‘Campaign_ID’\].nunique()}“)
print(f”Unique values in Location: {df\[‘Location’\].unique()}“)
print(f”Unique values in Customer_Segment:
{df\[‘Customer_Segment’\].unique()}“)
print(df\[‘Campaign_Type’\].value_counts())
print(df\[‘Channel_Used’\].value_counts()) #Exploratory Data Analysis
(EDA) and Visualization sns.set(style=”whitegrid”) #Campaign Performance
plt.figure(figsize=(8, 6)) sns.scatterplot(x=‘Acquisition_Cost’,
y=‘ROI’, data=df) plt.title(‘Relationship between Acquisition Cost and
ROI’) plt.show() plt.figure(figsize=(10, 6))
sns.barplot(x=‘Channel_Used’, y=‘Conversion_Rate’, hue=‘Campaign_Type’,
data=df) plt.title(‘Average Conversion Rate by Channel Used and Campaign
Type’) plt.show() plt.figure(figsize=(8, 6))
sns.boxplot(x=‘Campaign_Type’, y=‘Engagement_Score’, data=df)
plt.title(‘Distribution of Engagement Score by Campaign Type’)
plt.show() plt.figure(figsize=(10, 6)) sns.barplot(x=‘Company’, y=‘ROI’,
data=df) plt.title(‘Average ROI by Company’) plt.show()
plt.figure(figsize=(6, 4)) sns.heatmap(df\[\[‘Engagement_Score’,
‘Conversion_Rate’\]\].corr(), annot=True, cmap=‘coolwarm’)
plt.title(‘Correlation between Engagement Score and Conversion Rate’)
plt.show() #Customer Segmentation plt.figure(figsize=(8, 6))
sns.countplot(x=‘Target_Audience’, data=df) plt.title(‘Distribution of
Target Audience’) plt.show() plt.figure(figsize=(12, 6))
sns.barplot(x=‘Customer_Segment’, y=‘Conversion_Rate’, hue=‘Language’,
data=df) plt.title(‘Highest Conversion Rate by Customer Segment and
Language’) plt.show() plt.figure(figsize=(10, 6))
sns.boxplot(x=‘Customer_Segment’, y=‘Acquisition_Cost’,
hue=‘Channel_Used’, data=df) plt.title(‘Distribution of Acquisition Cost
by Customer Segment and Channel Used’) plt.show()
plt.figure(figsize=(10, 6)) sns.barplot(x=‘Language’,
y=‘Conversion_Rate’, data=df) plt.title(‘Average Conversion Rate by
Language’) plt.show() #Channel Effectiveness plt.figure(figsize=(10, 6))
sns.barplot(x=‘Channel_Used’, y=‘Engagement_Score’, hue=‘Campaign_Type’,
data=df) plt.title(‘Engagement Score by Channel Used and Campaign Type’)
plt.show() plt.figure(figsize=(8, 8))
df.groupby(‘Channel_Used’)\[‘ROI’\].sum().plot.pie(autopct=‘%1.1f%%’)
plt.title(‘Distribution of Total ROI by Channel Used’) plt.ylabel(’‘) \#
Hides the y-label plt.show() plt.figure(figsize=(8, 6))
sns.scatterplot(x=’Clicks’, y=‘Impressions’, hue=‘Channel_Used’,
data=df) plt.title(‘Relationship between Clicks and Impressions by
Channel Used’) plt.show() \# Time-Based Analysis plt.figure(figsize=(8,
6)) sns.histplot(df\[‘Duration’\], bins=20) plt.title(‘Distribution of
Campaign Duration’) plt.show() plt.figure(figsize=(12, 6))
sns.lineplot(x=‘Date’, y=‘Conversion_Rate’, hue=‘Company’, data=df)
plt.title(‘Trend of Conversion Rate over Date by Company’) plt.show()
plt.figure(figsize=(12, 6)) sns.lineplot(x=‘Date’, y=‘Engagement_Score’,
data=df) plt.title(‘Trend of Engagement Score over Date’) plt.show() \#
Geographic Analysis plt.figure(figsize=(10, 6))
sns.barplot(x=‘Location’, y=‘Acquisition_Cost’, data=df)
plt.title(‘Acquisition Cost by Location’) plt.show()
plt.figure(figsize=(12, 6)) sns.barplot(x=‘Location’,
y=‘Conversion_Rate’, hue=‘Target_Audience’, data=df)
plt.title(‘Conversion Rate by Location and Target Audience’) plt.show()
plt.figure(figsize=(8, 8))
df.groupby(‘Location’)\[‘ROI’\].sum().plot.pie(autopct=‘%1.1f%%’)
plt.title(‘Proportion of ROI by Location’) plt.ylabel(’’) plt.show()
